# **Modismos**

El término `modismo` surge en el ámbito de la programación en referencia a la guía de estilo enfocada a la construcción de código correcto, limpio y eficiente. En el ecosistema Python se conoce como programación "Phytonic". Veamos algunos ejemplos para entenderlo mejor: 

Si queremos crear una lista de 10 valores con tipo `NoneType` lo correcto será crear el código de manera compacta:

```
lista_nones = [None]*10
```

Para crear una lista con los 1000 primeros números naturales (el 0 es natural), una forma correcta será:

```
lista = [valor for valor in range(1000)]
```

Veámoslo ahora con un ejemplo en el que queremos filtrar la lista creada, quedándonos con los 600 primeros valores. Nótese que los siguientes códigos resultan ineficientes, al tener que recorrer 1000 veces el bucle en ambos casos para hallar la solución:

```
# Código ineficiente
for valor in lista:
  if valor >= 601:
    lista.remove(valor)
```

```
# Código ineficiente
while valor in lista:
  if valor >= 601:
    lista.remove(valor)
```

La forma correcta en este caso sería:

```
# Código eficiente
lista_filtrada = [valor for valor in lista if valor <= 600]
```

Veamos ahora un ejemplo en el que creamos una nueva lista a partir de otra, queriendo preservar la lista original:

```
# Código incorrecto
lista = [1 for valor in range(10)]
copia_seguridad_lista = lista # copia_seguridad_lista y lista quedean referidos al mismo objeto
for valor in range(len(lista)):
    lista[valor] *= 5 # Se destruye la copia de seguridad
print('lista =', lista)
print('copia_seguridad_lista =', copia_seguridad_lista)
```

```
# Código correcto
lista = [1 for valor in range(10)]
copia_seguridad_lista = lista # copia_seguridad_lista y lista quedean referidos al mismo objeto
lista = [valor * 5 for valor in lista] # Se preserva la copia de seguridad
print('lista =', lista)
print('copia_seguridad_lista =', copia_seguridad_lista)
```

Existen numerosos modismos en Python cuyo conocimiento es necesario para alcanzar un buen estilo de programación que conduzca al desarrollo de código correcto, eficiente y limipio. Recomendamos para ello la lectura de la "Guía de estilo en Python" de Guido van Rossum: documentos PEP 8 (https://peps.python.org/pep-0008/) y PEP 257 (https://peps.python.org/pep-0257/)
